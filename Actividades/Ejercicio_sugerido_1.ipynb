Usando el modelo de lazo cerrado de la Fig. 1-6, la función de trasferencia de lazo abierto G(s) específica de la pila del router (cuya longitud es la variable a controlar) denotado como Proceso es:

$G(s)=\frac{7031250 e^{-0.2s}}{(s+0.667)(s+5)(s+50)}$         (32)

La función $e^{-0.2s}$ representa el retardo. La función retardo se puede reemplazar por una función de aproximación racional como:

$e^{-sT} \approx \frac{1-\frac{T}{2} s}{1+\frac{T}{2}s}$     (33)

Una función de aproximación de Padé de primer orden que puede usarse es:

$e^{-sT} \approx 1-Ts$    (34)
